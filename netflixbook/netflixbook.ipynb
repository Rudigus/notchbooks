Primeiro, precisamos carregar o csv na memória (leitura)

```python
df_netflix = pd.read_csv("netflix_titles.csv")
```

Agora, vamos focar no Brasil

```python
df_netflix = df_netflix[df_netflix["country"] == "Brazil"].sort_values("release_year")
display(df_netflix)
```

Que tal ver que filmes da Anitta estão na Netflix?

```python
df_netflix = df_netflix[df_netflix["cast"].str.contains("Anitta", na = False)]
display(df_netflix)
```
